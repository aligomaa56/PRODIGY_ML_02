```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from matplotlib.colors import ListedColormap
```

```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
plt.scatter(df["Spending Score (1-100)"], df["Annual Income (k$)"])
plt.xlabel("Spending Score (1-100)")
plt.ylabel("Annual Income (k$)")
plt.title("Relationship Between Spending Score and Annual Income")
plt.grid(True)
plt.show()
```

```python
ssw = []
clus = range(1, 10)
for i in clus:
    ml = KMeans(n_clusters=i, init='k-means++', n_init=100)
    ml.fit(df[["Spending Score (1-100)", "Annual Income (k$)"]])
    ssw.append(ml.inertia_)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(clus, ssw, marker='o', linestyle='-', color='r')
plt.xlabel("Number of Clusters")
plt.ylabel("Sum of Squared Distances (SSW)")
plt.title("Elbow Method for Optimal K")
plt.grid(True)
plt.show()
```

```python
model = KMeans(n_clusters=5, init='k-means++', n_init=100)
df["y_hat"] = model.fit_predict(df[["Spending Score (1-100)", "Annual Income (k$)"]])

colors = ListedColormap(["#ff0000", "#00ff00", "#0000ff", "#ffff00", "#00ffff"])

plt.figure(figsize=(9, 7.5))
for i, cluster in df.groupby("y_hat"):
    plt.scatter(cluster["Spending Score (1-100)"], cluster["Annual Income (k$)"], color=colors(i), label=f"Cluster {i+1}")

plt.legend()
plt.title("Clustering of Customers based on Spending Score and Annual Income")
plt.xlabel("Spending Score (1-100)")
plt.ylabel("Annual Income (k$)")
plt.grid(True)
plt.show()
```
